```python
import pandas as pd
import numpy as np
#from re import sub
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.head())
customer_df.shape
```

```python
display(customer_df.info())
```

```python
#change titles to lower case 
snake = []
for i in customer_df.columns:
    snake.append(i.lower().replace(' ','_'))
customer_df.columns = snake

display(customer_df.head())

```

```python
# Change effective to date column to datetime format
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], errors='coerce')
display(customer_df)
```

```python
#Define a function that differentiates between continuous and discrete variables. 
#Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same 
#for discrete_df

#Look at some of the columns to get a feeling for the function:

print(customer_df['number_of_policies'].nunique())
print(customer_df['months_since_policy_inception'].nunique())
print(customer_df['number_of_open_complaints'].nunique())
print(customer_df['months_since_last_claim'].nunique())
```

```python
def different(x):
    categorical = x.select_dtypes(include = np.object)
    numerical = x.select_dtypes(include = np.number)
    
    discrete=[]
    continuous= []
    
    for column in numerical:
        if x[column].nunique() <= 10:
            discrete.append(x[column])
        else:
            continuous.append(x[column])
    
    discrete_df = pd.DataFrame(discrete).T
    continuous_df = pd.DataFrame(continuous).T
    
    return discrete_df, continuous_df, categorical
```

```python
discrete_df, continuous_df, categorical = different(customer_df)
```

```python
display(discrete_df)
display(continuous_df)
display(categorical)
```

```python
# Plot a correlation matrix, comment what you see.

correlations_matrix = continuous_df.corr()
fig, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

# Random variable distributions <a class="anchor" id="random_var"></a>

Insturctions

    Get the numerical variables from our dataset.
    Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
    Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
numericals = continuous_df
display(numericals)
```

```python
# Plot the distributions for the numerical data again:
plot_cont(numericals)
```

```python
#the distributions of customer_lifetime_value und monthly_premium_auto are looking like exponential distributions.
#total_claim_amount is a skewed distribution, that might become more normally distributed with a log transform

#Defining the log transform
def log_transfom_clean(x):
    if np.isfinite(x) and x!=0:
        return np.log10(x) 
    else:
        return np.NAN # We are returning NaNs so that we can replace them later


pd.Series(numericals['customer_lifetime_value'].apply(log_transfom_clean)).hist(bins=15)
plt.title('customer_lifetime_value')
plt.show()

pd.Series(numericals['monthly_premium_auto'].apply(log_transfom_clean)).hist()
plt.title('monthly_premium_auto')
plt.show()


# Customer Lifetime Value now looks a bit like a normal distribution but Monthly Premium Auto still looks like a exponential
# distribution
```

```python
pd.Series(numericals['total_claim_amount'].apply(log_transfom_clean)).hist(bins=15)
plt.title('total_claim_amount')
plt.show()
```

```python
#Total Claim Amount now is a bit skewed in the other direction, so the log transformation does not make a lot of sense here 
```
