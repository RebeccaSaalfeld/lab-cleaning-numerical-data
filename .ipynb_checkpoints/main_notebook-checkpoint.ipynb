```python
import pandas as pd
import numpy as np
#from re import sub
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.head())
customer_df.shape
```

```python
display(customer_df.info())
```

```python
#change titles to lower case 
snake = []
for i in customer_df.columns:
    snake.append(i.lower().replace(' ','_'))
customer_df.columns = snake

display(customer_df.head())

```

```python
# Change effective to date column to datetime format
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], errors='coerce')
display(customer_df)
```

```python
#Define a function that differentiates between continuous and discrete variables. 
#Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same 
#for discrete_df

#Look at some of the columns to get a feeling for the function:

print(customer_df['number_of_policies'].nunique())
print(customer_df['months_since_policy_inception'].nunique())
print(customer_df['number_of_open_complaints'].nunique())
print(customer_df['months_since_last_claim'].nunique())
```

```python
def different(x):
    categorical = x.select_dtypes(include = np.object)
    numerical = x.select_dtypes(include = np.number)
    
    discrete=[]
    continuous= []
    
    for column in numerical:
        if x[column].nunique() <= 10:
            discrete.append(x[column])
        else:
            continuous.append(x[column])
    
    discrete_df = pd.DataFrame(discrete).T
    continuous_df = pd.DataFrame(continuous).T
    
    return discrete_df, continuous_df, categorical
```

```python
discrete_df, continuous_df, categorical = different(customer_df)
```

```python
display(discrete_df)
display(continuous_df)
display(categorical)
```

```python
# Plot a correlation matrix, comment what you see.

correlations_matrix = continuous_df.corr()
fig, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python

#plot continuous variables:
def plot_cont(x):
        for column in x.columns:
            plt.title(column)
            x[column].hist(bins=25)
            plt.show()
            

#plot discrete variables:
def plot_disc(x):
        for column in x.columns:
            plt.title(column)
            x[column].hist(bins=25)
            plt.show()
```

```python
#Look for outliers in the continuous variables. (HINT: There’s a good plot to do that!)


def box_plot(x):
    for column in x.columns:
        sns.boxplot(y=x[column])
        plt.title(column, fontsize=16)
        plt.show()

box_plot(continuous_df)
```
